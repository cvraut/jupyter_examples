# using varcode

```python
import os
os.environ['PYENSEMBL_CACHE_DIR'] = '/nfs/corenfs/INTMED-Speliotes-data/External_Data/PYENSEMBL_CACHE/'
from varcode import Variant
from pyensembl import ensembl_grch38
```

```python
#PNPLA3 rs738409 https://www.ncbi.nlm.nih.gov/snp/rs738409
myVariant = Variant(contig=22, start=43928847, ref="C", alt="G", ensembl=ensembl_grch38)
```

```python
myVariant.coding_genes
```

```python
myEffects = myVariant.effects()
myEffects
```

```python
topPriorityEffect = myEffects.top_priority_effect()
topPriorityEffect
```
