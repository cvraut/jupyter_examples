# whecho tutorial

Recently, Wendy and I came up with a tool that I thought would be useful to keep track of long-running jobs. We titled it [whecho (webhooks echo)](https://github.com/cvraut/whecho).



The goal is that it should work like linux echo except instead of printing the echo'ed string to stdout it prints it to the webhook that the user sets.

# installation

whecho is already setup on [speliotes-jupyterhub](https://speliotes-jupyter.med.umich.edu/)

If you want to set it up yourself, in a conda environment u must run `pip install whecho`. (Note: this will only install whecho in the current conda environment and will have to repeated if you use multiple conda environments frequently). It is unadvised to install whecho globally as it is written in python and might clash with loaded python versions.

# First time setup

If you have never used whecho, it is advised to run `whecho --init` and set the default url to channel that you are subscribed to.

To start with feel free to use [the webhook created in our servers channel](https://discord.com/channels/1073360774615011348/1157410255181987851/1201585159368089641) (access restricted for obvious reasons):


But for prolonged or advanced usage I would recommend setting up a private server and creating a private webhook url for a channel in that server.
